```python
def dep():
    a1 = int(input("Enter deposit amount: "))
    amount.append(a1)
    print(f"Deposited: {a1}")

def wihdr():
    current_balance = sum(amount) 
    a2 = int(input("Enter withdrawal amount: "))
    
    if a2 > current_balance:
        print(f"Not sufficient funds. Your balance is only: {current_balance}")
    else:
        amount.append(-a2)
        print("Withdrawal successful")

def bal():
    total = sum(amount) 
    print(f"Current Balance: {total}")

1
```
