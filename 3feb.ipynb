```python
def calc22(*args):
    ssum=0
    for i in args:
        ssum+=i
    return ssum
```

```python
calc22(7,8,9,4,6,2,5)
```

```python
def calc23(*args):
    ssum=0
    length = 0                     #average
    for i in args:
        ssum+=i
    length=len(args)
    return ssum/length
```

```python
calc23(7,8,9,14,6,2,5)
```

```python
def add(a,b):
    print (a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
    
```

```python
print("calculator")
choice = input("enter choice").lower()
num1 = int(input("num1"))
num2 = int(input('num2'))
if choice=='add':
    add(num1,num2)
elif choice=='sub':
    sub(num1,num2)
elif choice=='mul':
    mul(num1,num2)
else:
    print('wrong choice')
```
